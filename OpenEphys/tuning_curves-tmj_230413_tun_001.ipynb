```python
trial_spikes = tuning.get_trial_spikes(cluster_spike_times, stim_times, tuning_param)
print(f"{trial_spikes.shape[0]} out of {cluster_spike_times.shape[0]} total spikes {trial_spikes.shape[0]/cluster_spike_times.shape[0]}")
```

```python
# calculate histogram across all trials

start = tuning_param['pre_window']
stop = tuning_param['max_trial_duration']
step = tuning_param['bin_size']
bins = np.arange(start, stop, step)
hist_y, hist_x = tuning.psth(trial_spikes[:,2],bins)
plt.plot(hist_x[0:-1], hist_y)
plt.xlabel("time (s)")
plt.ylabel("response (spikes)")
plt.savefig(python_figures + 'cluster_' + str(current_cluster_number) + '_psth.pdf')
```

```python
# calculate histogram for each trial
start = tuning_param['pre_window']
stop = tuning_param['max_trial_duration']
step = tuning_param['bin_size']
bins = np.arange(start, stop, step)
trial_psth =  np.full((stim_times.shape[0], bins.shape[0]-1), np.nan, dtype = float)
for trial_index in range(stim_times.shape[0]):
    current_spikes = trial_spikes[:,1] == trial_index
    current_spikes = trial_spikes[(current_spikes),2]
    hist_y, hist_x = tuning.psth(current_spikes,bins)
    trial_psth[trial_index,:] = hist_y


# plt.plot(hist_x[0:-1], hist_y)
# plt.xlabel("time (s)")
# plt.ylabel("response (spikes)")
# plt.savefig("D:\\AwakeData\\Python\\230413\\smr_files\\tmj_230413_ori_003\\figures\\all_spikes_psth.pdf")
```

```python
plt.plot(np.mean(trial_psth, axis = 0) / tuning_param['bin_size'])
```

```python
z = bins[0:-1]>0
trial_mean = np.sum(trial_psth[:,(z)], axis = 1)
print(f"# of trials = {trial_psth.shape[0]}")
trial_mean.shape
iv_unique = np.unique(IV)
print(f"IV Count  = {len(iv_unique)}")
m = np.full((len(iv_unique)),np.nan, dtype = float)
for iv_index in range(len(iv_unique)):
    current_trials = IV == iv_unique[iv_index]
    trial_mean[current_trials]
    m[iv_index] =  np.mean(trial_mean[current_trials])

plt.plot(iv_unique, m)

plt.xlabel("ori(" + chr(176) +")")
plt.ylabel("response (spikes)")
plt.savefig(python_figures  + 'cluster_' + str(current_cluster_number) +  '_ori_tuning.pdf')
```

```python
# cycle histogram


stim_times, IV = OpenEphys.get_synced_stim_info(stim_info_file)

#create copy of stim time for modification to make cycle hists
cycle_stim_times = stim_times.copy()

cycle_duration =  1/tuning_param['TF']

#divide spikes into trials and reset spike times relative to stim onset
# change trigger times so that the trial ends at a multiple of a full cycle
cycle_stim_times = tuning.cycle_match(cycle_stim_times, cycle_duration)
cycle_spikes = tuning.get_trial_spikes(cluster_spike_times, cycle_stim_times, tuning_param)


cycle_hist = tuning.create_cycle_hist(tuning_param, stim_times, cycle_spikes)

plt.plot(np.mean(cycle_hist, axis = 0)/ tuning_param['bin_size'])

# fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)
# fig.tight_layout()
```
